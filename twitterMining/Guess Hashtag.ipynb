```python
def file_to_list(file_name):
    """
    file_to_list reads a file and converts each line to list items
    
    Args:
        file_name (string): the name of the file
        
    Returns:
        (list): the items that are in the file
    """
    results = []
    with open(file_name, 'rt') as f:
        for line in f:
            results.append(line.replace('\n', ''))
    return results
```

```python
tweets = file_to_list('hillaryTrump.txt')
print(len(tweets))
```

```python
hillary_tweets = []
trump_tweets = []
neither = []
for tweet in tweets:
    lowercase_tweet = tweet.lower()
    
    if '#hillary' in lowercase_tweet:
        hillary_tweets.append(lowercase_tweet)
    if '#trump' in lowercase_tweet:
        trump_tweets.append(lowercase_tweet)
    if '#hillary' not in lowercase_tweet and '#trump' not in lowercase_tweet:
        neither.append(lowercase_tweet)
            
print(len(tweets))
print(len(hillary_tweets))
print(len(trump_tweets))

print(len(hillary_tweets) + len(trump_tweets))
```
